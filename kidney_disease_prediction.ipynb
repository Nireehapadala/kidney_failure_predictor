```python
import keras
```

```python
from keras.models import Sequential
obj= Sequential()
```

```python
from keras.layers import Dense
```

```python
import pandas as pd
```

```python
dt=pd.read_csv('/content/kidney_disease.csv')
dt.head()
```

```python
dt = dt.dropna()
dt.head(20)
```

```python
mapping = {
    'yes': 1, 'no': 0,
    'normal': 1, 'abnormal': 0,
    'present': 1, 'notpresent': 0,
    'good': 1, 'poor': 0,
    'ckd': 1, 'notckd': 0
}
def convert_column(col):
    return col.map(lambda x: mapping.get(str(x).strip().lower(), x))
dt= dt.apply(convert_column)
dt= dt.apply(pd.to_numeric, errors='ignore')
dt.head()
```

```python
dt.head(50)
```

```python
dt.corr()
```

```python
dt=dt.drop(['bp','ba'],axis=1)
```

```python
dt.tail(2)
```

```python
x=dt.iloc[:,0:23]
y=dt['classification']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
print(x.shape,y.shape,x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
obj.add(Dense(units=23,activation='relu'))
obj.add(Dense(units=5,activation='relu'))
obj.add(Dense(units=1,activation='sigmoid'))
```

```python
obj.build(input_shape=(None,23))
```

```python
obj.summary()
```

```python
obj.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
obj.fit(x_train,y_train,epochs=100,batch_size=8,steps_per_epoch=x_train.shape[0]//8)
```

```python
y_pred=obj.predict(x_test)
```

```python
y_pred=(y_pred>0.5).astype(int)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
obj.save('kidney.h5')
```
